```python
#Combien y-a-t-il de musées dont le nom commence par "Château" ?
# get list of all museum names - done
# check each element in nameslist if musseum starts with chateau
#get the length of the list

def countName():
    names = (spreadsheet["name"])
    castleNames = names.str.startswith("Chapelle")
    print(names)
    #print("Names containing Chateau:",castleNames)
    count=0
    for elt in castleNames:
        if elt == True:
            count +=1
    #liste = (spreadsheet["name"])
    print("Total number of museum containing:",count)

    
countName() 
count=0
#print("Total number of museum containing Chapelle in their name",count)
    
```

```python
#Combien de musées ont "Napoléon" dans leur nom ?


def countNapoleon(count):
    names = (spreadsheet["name"])
    castleNames = names.str.contains("Napoleon")
    #print(names)
    #print("Names containing Napoleon:",castleNames)
    count=0
    for elt in castleNames:
        if elt == True:
            count +=1
    #liste = (spreadsheet["name"])
    print("Total number of museum containing Chapelle in their name",count)

    
countNapoleon(count) 
count=0
#print("Total number of museum containing Napoleon in their name",count)
  
```

```python
#Quel est le nombre moyen de musées par ville ?
#length of city column
#length of name column
#calculate average
museums = (spreadsheet["name"])
cities = (spreadsheet["city"]).drop_duplicates()

print("number of museums",len(museums))
numberMuseums = len(museums)
print("number of cities", len(cities))
numberCities = len(cities)

average = numberMuseums/numberCities
print("Average",average)
```

```python
#Dans quelle(s) ville(s) y-a-t-il de plus de musées ?
#pivot table ligne cities , summe musee comme colonne

table = spreadsheet.pivot_table("name",index="city",aggfunc="count")
#print(table.max())
print(table)
sortedTable = table.sort_values(by="name",ascending=True)
print("City with most museums:",sortedTable.tail(1))
```

```python
#Quel est le nombre médian de musées par ville ?
#DataFrame.median(axis=_NoDefault.no_default, skipna=True, level=None, numeric_only=None, **kwargs)[source]
##median = spreadsheet.median(axis=1, skipna=True)
table = spreadsheet.pivot_table("name",index="city",aggfunc="count")
cities = table.median()
print("Median:",cities)
#cities = table["name"]
#print(cities)
```
